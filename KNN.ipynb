```python
rawdata['Rating'] =rawdata['Rating'].fillna(0) 
rawdata['REVIEWS'] =rawdata['REVIEWS'].fillna(0)
rawdata['SIZE'] =rawdata['SIZE'].fillna(0)
```

```python
selected_movie = rawdata[rawdata['App']=="FR Calculator"].iloc[0]
distance_column = ['Rating','REVIEWS','SIZE']
```

```python
def euclidean_distance(row):
  inner_value = 0
  for k in distance_column:
    inner_value += (row[k] - selected_movie[k]) ** 2
  return np.sqrt(inner_value)

movie_distance = rawdata.apply(euclidean_distance,axis = 1)
```

```python
ds_numeric = rawdata[distance_column]
ds_normalized = (ds_numeric - ds_numeric.mean())/ds_numeric.std()
```

```python
from scipy.spatial import distance
movie_normalized = ds_normalized[rawdata['App']=="FR Calculator"]
#Find Distance
euclidean_distance = ds_normalized.apply(lambda row: distance.euclidean(row, movie_normalized), axis=1)
#Create a new database with distances
distance_frame = pd.DataFrame(data={"dist": euclidean_distance, "idx": euclidean_distance.index})
distance_frame.sort_values(by = "dist")

#Find the most similar movie to Fuzzbucket
smallest = distance_frame.iloc[1]["idx"]
most_similar = rawdata.loc[int(smallest)]["App"]
```

```python
import random
from numpy.random import permutation
#Random shuffle the index of df
random_indices = permutation(rawdata.index)

#set a cutoff for how many items we want 
test_cutoff = np.floor(len(rawdata)/3)
 #generate the test set
test = rawdata.loc[random_indices[1:int(test_cutoff)]]

 #generate the train set
train = rawdata.loc[random_indices[int(test_cutoff):]]
```

```python
x_columns = ['Rating','REVIEWS','SIZE']
y_column = ['Rating']

from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(train[x_columns],train[y_column])
predictions = knn.predict(test[x_columns])
print(predictions)
```
