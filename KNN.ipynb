## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

## Importing the dataset

```python
from google.colab import drive
drive.mount('/content/drive')

dataset = pd.read_csv('/content/drive/MyDrive/6th Sem/Machine Learning/Social_Network_Ads - Social_Network_Ads.csv')
X = dataset.iloc[:, [1, 2, 3]].values
y = dataset.iloc[:, -1].values
```

```python
dataset.head(10)
```

**`Encode character values`**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,0] = le.fit_transform(X[:,0])
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the K-NN model on the Training set

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train) 
```

**# Predict output for the test set and compare with the existing Y value**

```python
y_pred = classifier.predict(X_test)
print (y_pred)
print (y_test)
```

**#Accuracy and Confusion Matrix values**

```python
from sklearn.metrics import confusion_matrix,accuracy_score

cm=confusion_matrix(y_test,y_pred)
print (cm)

ac = accuracy_score(y_test,y_pred)
print (ac)
```

## Predicting a new result

```python
print(classifier.predict(sc.transform([[1,30,870000]])))
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
